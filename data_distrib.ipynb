Вивчення структури даних

завантажимо все в ДФ та подивимось що в нас є (також додано файл data.xls де можно в Екселі подивитись на дані )

```python
import pandas as pd

# Шлях до вашого CSV файлу
file_path = 'internet_service_churn.csv'

# Завантаження CSV файлу в DataFrame
df = pd.read_csv(file_path)

# Перевірка перших кількох рядків DataFrame
print(df.head())
```

```python
# Виведення типів даних кожної колонки
print("Типи даних у кожній колонці:")
print(df.dtypes)
```

```python
# Виведення детальної інформації про DataFrame
print("Інформація про DataFrame:")
df.info()
```

Є незаповненні значення ,а також типи даних або флоат або інт

Подивимось що в колонках int

```python
# Вибір тільки колонок з типом даних 'int'
int_columns = df.select_dtypes(include='int')

# Перевірка унікальних значень в кожній колонці з типом даних 'int'
unique_values = {column: int_columns[column].unique() for column in int_columns.columns}

# Виведення унікальних значень
for column, values in unique_values.items():
    print(f"Унікальні значення в колонці '{column}':")
    print(values)
    print("\n")
```

Колонку ID далі не разглядаємо, там унікальний номер абоненту

### Подивимось що у колонках флоат

```python
# Вибір тільки колонок з типом даних 'float'
float_columns = df.select_dtypes(include='float')

# Перевірка унікальних значень в кожній колонці з типом даних 'float'
unique_values = {column: float_columns[column].unique() for column in float_columns.columns}

# Виведення унікальних значень
for column, values in unique_values.items():
    print(f"Унікальні значення в колонці '{column}':")
    print(values)
    print("\n")
```

```python
# Отримання мінімальних та максимальних значень для кожної колонки
min_values = df.min()
max_values = df.max()

# Виведення мінімальних та максимальних значень
for column in df.columns:
    print(f"Колонка '{column}':")
    print(f"  Мінімальне значення: {min_values[column]}")
    print(f"  Максимальне значення: {max_values[column]}")
    print("\n")
```

### маштаб різний (бінарні в розгляд поки не беремо)

також в колонці 'subscription_age' є від'ємне значення і воно тільки одне ,я думаю ,що це помилка ,пропоную змінити на + 

```python
# Заміна всіх відсутніх значень на 0
df_filled = df.fillna(0)
```

Замінив всі НаН на 0

Приберемо колонки ,які на мій погляд поки не потрібні для аналізу

це

'id'
'is_tv_subscriber'
'is_movie_package_subscriber'
'churn'

```python
# Перелік колонок, які потрібно видалити
columns_to_drop = ['id', 'is_tv_subscriber', 'is_movie_package_subscriber', 'churn']

# Видалення колонок
df_cleaned = df_filled.drop(columns=columns_to_drop)

# Перевірка результату
print(df_cleaned.head())

```

```python
# Описова статистика
print(df_cleaned.describe())
```

```python
import matplotlib.pyplot as plt

# Створення гістограм для кожної колонки
df_cleaned.hist(bins=60, figsize=(15, 10))
plt.show()
```

```python
import seaborn as sns

# Створення теплової карти кореляції
plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
